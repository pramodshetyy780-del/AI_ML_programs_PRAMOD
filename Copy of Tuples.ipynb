```python
x=("Hari","Potato","Honey")
print(x)
y=list(x)
print(y)
y[1]="Kiwi"
x=tuple(y)
print(x)

#appending in tuple
thistuple=("Hari","Potato","Honey")
y=list(thistuple)
y.append("No More")
thistuple=tuple(y)
print(thistuple)
```

```python
#remove items from tuple
thistuple = ("Hari","Potato","Honey") # Re-initialize thistuple to ensure 'Hari' is present
y=list(thistuple)
y.remove("Hari")
thistuple=tuple(y)
print(thistuple)

#delete tuple completely
thistuple=("Hari","Potato","Honey")
del thistuple
#print(thistuple)
```

```python
#pack
fruits=("Apple","Banana","Melons")
print(fruits)

#unpacking
(red,green,blue)=fruits
print(red)
print(blue)
print(green)
```

```python
citric_fruits = ("orange", "lemon", "grapefruits","tangerines")
(yellow,*green,red)=citric_fruits
print(yellow)
print(green)
print(red)
```
