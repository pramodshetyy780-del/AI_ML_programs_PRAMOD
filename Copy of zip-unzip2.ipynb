```python
#combining enume*rate and zip
crops=["rice","wheat","ragi"]
yields=[1200,1500,110]
for plot_no,(crop,crop_yield) in enumerate(zip(crops,yields)):
  print(f"plot{plot_no}: {crop}-{crop_yield}kg")
```

```python
records = [
    (1, "Rice" ),
    (2, "Wheat" ),
    (3, "Maize" )
]

plot_no, crop = zip(*records)

print(plot_no)
print(crop)
```

```python
#Zip two dictionaries (DEFAULT behavior)
#By default, zip() iterates over dictionary keys

crop_price = {
    "Rice": 40,
    "Wheat": 35,
    "Maize": 30
}

crop_yield = {
    "Rice": 1200,
    "Wheat": 1500,
    "Maize": 1100
}

for a, b in zip(crop_price.keys(), crop_yield.values()):
    print(a, b)
```

```python
for crop,price,yield_kg in zip(crop_price.keys(),crop_price.values(),crop_yield.values()):
  print(f"{crop}:price={price},yield={yield_kg}")
```

```python
items = list(crop_price.items())
keys, values = zip(*items)
print(crop_price.items())
print(keys)
print(values)
```
