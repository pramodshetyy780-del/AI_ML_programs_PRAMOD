```python
def crop_dict(filename):
 crops=[]
 with open(filename) as file:
   for line in file:
    data=line.split(",")
    if line:
      crop={
          "cropname":data[0],
          "season":data[1],
          "area":float(data[2]),
          "yield":float(data[3]),
          "cost":float(data[4]),
          "returns":float(data[5])
      }
      crops.append(crop)
 return crops
crop_dict("data_set.txt")
```

```python
def find_kharif_crop(crops):
  kharif_crop=[]
  non_kharif=[]
  for crop in crops:
    if crop["season"]=="kharif":
     kharif_crop.append(crop["cropname"])
    else:
     non_kharif.append(crop["cropname"])
  return kharif_crop,non_kharif
  crops=crop_dict("data_set.txt")
find_kharif_crop(crops)
```

```python
def crop_profit(crops):
  for crop in crops:
    profit=crop["returns"]-crop["cost"]
    print(crop["cropname"],"profit=",profit)
crop_profit(crops)
```

```python
def summary(crops):
  crop_area=[]
  total_yield=[]
  for crop in crops:
   crop_area.append(crop["area"])
   total_yield.append(crop["yield"])
   total_area=sum(crop_area)
  calculated_yield=sum(total_yield)
  avg_yield=calculated_yield/len(crops)
  print("total_area=",total_area,",average_yield=",avg_yield)
summary(crops)
```
