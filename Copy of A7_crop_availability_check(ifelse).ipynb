Lab Assignment 7: Crop Availability Check
Given the crop list:
["Rice", "Wheat", "Maize"]
Accept a crop name from the user and:
Display "Crop available" if the crop is present in the list
Otherwise, display "Crop not available"



```python
crop_name = str(input("Enter CROP NAME:\t"))
crop_list = ["Rice", "Wheat", "Maize"]
if crop_name in crop_list:
  print("Crop available")
else:
  print("Crop not available")
```
