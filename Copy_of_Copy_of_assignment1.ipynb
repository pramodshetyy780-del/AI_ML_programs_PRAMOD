```python
print("===suitability of rice===")
rainfall=float(input("enter the the rainfall:"))
soil_moisture=int(input("enter the soil moiture percentage:"))
if soil_moisture>=70 and rainfall>=300:
  print("suitable for rice")
else:
  print("not suitable for rice")
```

```python
print("===fertilizer recommendation system===")
crop_name=input("enter the crop name(rice/wheat/maize):")
nitrogen_level=int(input("enter the nitrogen level in soil:"))
if crop_name=="rice" and nitrogen_level<50:
  print("apply urea")
elif crop_name=="wheat" and nitrogen_level<40:
  print("apply dap")
else:
  print("fertilizer not required")
```

```python
print("===crop yeild classification===")
cropyield=float(input("enter the crop yield in quintals:"))
if cropyield>=50:
  print("highyield")
elif cropyield<30:
  print("low yeild")
else:
  print("medium yeild")
```
