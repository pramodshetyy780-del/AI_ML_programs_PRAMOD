Lab Assignment 9: Seed Quality Evaluation
Accept the following inputs from the user:
Germination percentage
Moisture content
Apply the following rules:
If germination percentage is greater than or equal to 85:
If moisture content is less than or equal to 12, display "High quality seeds"
Otherwise, display "Moisture too high"
If germination percentage is less than 85, display "Low germination"


```python
seed_germination = int(input("Enter the seed germination percentage: \t"))
seed_moisture = int(input("Enter the seed moisture content: \t"))
if seed_germination >= 85 and seed_moisture <= 12:
  print("High quality seeds")
elif seed_germination >= 85 and seed_moisture > 12:
  print("Moisture too high")
elif seed_germination < 85:
  print("Low germination")
```
