Lab Assignment 8: Average Rainfall Analysis
Accept rainfall values for 5 days from the user and store them in a list.
Calculate the average rainfall
If the average rainfall is greater than or equal to 100, display "Sufficient rainfall"
Otherwise, display "Insufficient rainfall"


```python
rf_d1=float(input("Enter rainfall for day 1: \t"))
rf_d2=float(input("Enter rainfall for day 2: \t"))
rf_d3=float(input("Enter rainfall for day 3: \t"))
rf_d4=float(input("Enter rainfall for day 4: \t"))
rf_d5=float(input("Enter rainfall for day 5: \t"))
rf_list=[rf_d1,rf_d2,rf_d3,rf_d4,rf_d5]
avg_rf=sum(rf_list)/5
print("Average rainfall is: \t",avg_rf)
if avg_rf>=100:
  print("Sufficient rainfall")
else:
  print("Insufficient rainfall")
```
