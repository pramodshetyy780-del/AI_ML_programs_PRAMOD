```python
#increase yield due to fertilizer usage
def increase_yield(yields,increment):
  for i in range(len(yields)):
    yields[i]+=increment

cropyields=[1200,1350,1100]
increase_yield(cropyields,100)
print(cropyields)
```

```python
#convert to quintals
def convert_to_quintals(yields):
  result=[]
  for value in yields:
        result.append(value/100)
  return result
crop_yields=[1200,1350,1100]
quintals=convert_to_quintals(crop_yields)
print(quintals)
```

```python
def high_yld_crp(yld,th):
  high=[]
  for value in yld:
   if value>=th:
    high.append(value)
  return high

crp_yld=[1200,900,1500,80]
print(high_yld_crp(crp_yld,1000))
```
