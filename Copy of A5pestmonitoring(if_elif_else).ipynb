Lab Assignment 5: Pest Monitoring System
Repeatedly accept the pest count from the user.
Continue accepting values until the pest count becomes 0
When pest count is 0, display "Monitoring stopped"

```python
num_pest=int(input("Enter the number of pests:\t "))
if num_pest>=1:
  print("Monitoring Needed, kindly follow IPM")
elif num_pest==0:
  print("Monitoring stopped")
else:
  print("Invalid input")
```
