Lab Assignment 10: Harvest Readiness Monitoring
Repeatedly accept the crop maturity percentage from the user.
Continue until maturity percentage becomes 100 or more
Display "Crop ready for harvest"

```python
crop_matuity=float(input("Enter the crop maturity percentage: \t"))
while crop_matuity<100:
    crop_matuity=float(input("Enter the crop maturity percentage: \t"))
    if crop_matuity>=100:
        print("Crop ready for harvest")
```
