```python
li=[1,1.33,'gfg',0,'no',None,'g',True]
val1,val2=0,''
for x in li:
    if(type(x)==int or type(x)==float):
      val1+=x
    elif(type(x)==str):
      val2+=x
    else:
     break
print(val1,val2)
```
