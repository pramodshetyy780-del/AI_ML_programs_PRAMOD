```python
import numpy as np
yields=np.array([
    [45,48,52,49,53],
    [38,41,39,44,42],
    [52,55,58,56,60],
    [41,43,45,47,46]
])
weather_factors=np.array([1.0,0.95,1.05,0.98,1.02])
print('original yields(tons/hectare):')
print(yields)
print('\n weather impact factors:',weather_factors)
adjusted_yields=yields*weather_factors
print('\n weather_adjusted yields:')
print(adjusted_yields.round(2))
exceptional=adjusted_yields>55
print('\nexceptional yields(>55):')
print(adjusted_yields[exceptional])
categories=np.where(
    adjusted_yields>50,'high',
    np.where(adjusted_yields>40,'medium','low')
)
print('\nyield categories:')
print(categories)
selected_farms=[0,2]
print('\nselected farm yields:')
print(yields[selected_farms])
good_years=adjusted_yields[(adjusted_yields>45)&(adjusted_yields<55)]
print(f'\ngood performing years(45-55 tons):')
print(good_years)
```
