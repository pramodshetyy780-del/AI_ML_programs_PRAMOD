```python
number_of_fields=5
area_of_fields=2.5
crop_name="rice"
irrigation_available=False
print("Crop name:",crop_name)
print("number of fields:",number_of_fields)
print("area per field:",area_of_fields,"acres")
print("irrigation available:",irrigation_available)
total_area=number_of_fields*area_of_fields
print("total cultivated area:",total_area,"acres")
if irrigation_available:
  print("status:irrigation facility is available")
else:
  print("status:no irrigation facility")
```
