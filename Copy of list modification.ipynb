Range of Indices

```python
thislist=["apple","banana","cherry","orange","kiwi","melon","mango"]
print(thislist[2:5])
```

```python
thislist=["apple","banana","cherry","orange","kiwi","melon","mango"]
print(thislist[:4])
```

```python
thislist=["apple","banana","cherry","orange","kiwi","melon","mango"]
print(thislist[-4:-1])
```

```python
thislist=["apple","banana","cherry"]
thislist[1] ="blackcurrent"
print(thislist)
```

```python
thislist=["apple","banana","cherry","mango","jackfruit"]
thislist[0:4]=["blackcurrent","watermelon"]
print(thislist)
```

```python
thislist=["apple","banana","cherry"]
thislist.insert(2,"watermelon")
print(thislist)
```

```python
thislist=["apple","banana","cherry"]
thislist.append("watermelonOrange")
print(thislist)
```

```python
thislist=["apple","banana","cherry"]
tropical=["Papaya","guava","mango"]
thislist.extend(tropical)
print(thislist)
```

```python
thislist=["apple","watermelon","banana","cherry"]
thislist.remove("watermelon")
print(thislist)
```

```python
thislist=["apple","watermelon","banana","cherry"]
thislist.pop(1)
print(thislist)
```

```python
thislist=["apple","watermelon","banana","cherry","banana"]
for item in thislist:
  if  item=="banana":
    thislist.remove("banana")
print(thislist)
```
