```python
plots = []
crops = []
yields = []

# Get the number of entries from the user once
try:
    n = int(input("Enter the number of plots/entries: "))
except ValueError:
    print("Invalid input. Please enter an integer.")
    n = 0 # Default to 0 to prevent further execution if input is invalid

if n > 0:
    # Collect plot numbers
    for i in range(n):
        plot_number = input(f"Enter plot number {i+1}: ")
        plots.append(plot_number)
    print("Collected plot numbers:", plots)

    # Collect crop names
    for i in range(n):
        crop_name = input(f"Enter Crop name for plot {plots[i]}: ")
        crops.append(crop_name)
    print("Crops grown :", crops)

    # Collect yields
    for i in range(n):
        yield_kg = input(f"Enter yield in kg for plot {plots[i]} ({crops[i]}): ")
        yields.append(yield_kg)
    print("Collected yields:", yields)

    # Iterate and print combined information
    # Using 'plot_val', 'crop_val', 'yield_val' as loop variables to avoid shadowing the lists
    for plot_val, crop_val, yield_val in zip(*(plots, crops, yields)):
        print(f"Plot {plot_val} grows {crop_val} with yield {yield_val} kg")
else:
    print("No entries to process.")
```
